```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
site = 'https://ballotpedia.org/United_States_Congressional_elections_results,_2012#US_House_Margin_of_Victory'
```

```python
response = requests.get(site)
response.status_code
page = response.text
```

```python
margin2012 = BeautifulSoup(page, 'html.parser')
```

```python
tables = margin2012.find_all('table')
```

```python
margtab = tables[3]
```

```python
margtab
```

```python
rows = [rows for rows in margtab.find_all('tr')]
```

```python
rows[2]
```

```python
top = rows[2]
```

```python
b = top.find_all('td')
```

```python
districts = {}
for row in rows[2:-1]:
    items = row.find_all('td')
    district = items[0].text
    links = items[1].find_all('a')
    party = links[0].get('title')
    name = links[1].get('title')
    districts[district] = [party] + [name] + [i.text for i in items[2:]]
     

```
